# Step 2: Exploration, Cleaning & Preprocessing

# categorical variables

```python
# Merge the categories other and unknown within poutcome as they are the same thing
df["poutcome"] =  df["poutcome"].replace("other", "unknown")
```

```python
# Separate the categorical variables except for the outcome variable "deposit"
c = df[["job", "marital", "education", "default", "housing", "loan", "contact", "month", "poutcome"]]
```

```python
# Investigate distribution of target variable (deposit)
plt.figure(figsize = (8,8))

plt.pie(df['deposit'].value_counts(),
        colors = sns.color_palette('pastel'),
        labels = df['deposit'].value_counts().index,
        autopct = '%.0f%%',
        explode = (0, 0.03)
        )

plt.title("Was the telemarketing campaign successful?")
```

```python
# Create countplot of each variable

plt.figure(figsize = (20,20))

for index, element in enumerate(c):
  plt.subplot(5,2,index+1)
  sns.countplot(data=c,
                y=element,
                order=c[element].value_counts().index,
                hue=element,
                palette="pastel",
                )
  plt.xlabel("")
  plt.ylabel("")
  plt.title(f"Number of observations within categorical variable {element}")
  plt.tight_layout()
```

```python
# Create countplot of each variable, split by target variable "default"

plt.figure(figsize = (20,20))

for index, element in enumerate(c):
  plt.subplot(5,2,index+1)
  sns.countplot(data = c,
                y = element,
                hue = df["deposit"],
                order = c[element].value_counts().index,
                palette = "pastel",
                )
  plt.xlabel("Number of observations")
  plt.ylabel("")
  plt.title(f"Number of observations within {element}, split by outcome of campaign")
  plt.tight_layout()
```

```python
# Create 100% stacked bar chart for each variable, with categories of bar adding up to 100% within each bar

fig, axes = plt.subplots(5, 2, figsize=(20, 20)) # create matrix for plots
axes = axes.flatten() # convert into 1D array for easier indexing

for index, element in enumerate(c):
  data = pd.crosstab(c[element], df["deposit"], normalize = "index") # calculate proportions
  data = data[["yes", "no"]] # put yes first
  data = data.sort_values("yes", ascending=True) # sort by yes
  data.plot(kind="barh",
            stacked=True,
            ax = axes[index],
            color = sns.color_palette("pastel")
            )

  axes[index].set_title(f"Proportion of converted customers within each group of {element}")
  order = data.value_counts().index
```

# quantitative variables

```python
# Violin plots for checking the distribution:

plt.figure(figsize = (20,20))

for index, element in enumerate(q):
  plt.subplot(4,3,index+1)
  sns.violinplot(data = q,
                 y=element,
                 color = "#A8C8E8"
                )
  plt.xlabel("")
  plt.ylabel("")
  plt.title(f"Distribution quantitative variable {element}")
  plt.tight_layout()
```

```python
# Violin plots for checking the distribution, split by the outcome variable "deposit":

plt.figure(figsize = (20,20))

custom_colors = {"yes": "#A8C8E8", "no": "#FFB07C"}

for index, element in enumerate(q):
  plt.subplot(4,3,index+1)
  sns.violinplot(data = q,
                 y=element,
                 hue = df["deposit"],
                 palette = custom_colors
                )
  plt.xlabel("")
  plt.ylabel("")
  plt.title(f"Distribution quantitative variable {element}")
  plt.tight_layout()
```

```python
# Boxplots for checking outliers:

plt.figure(figsize = (20,20))

for index, element in enumerate(q):
  plt.subplot(4,3,index+1)
  sns.boxplot(data = q,
              y=element,
              color = "#A8C8E8"
              )
  plt.xlabel("")
  plt.ylabel("")
  plt.title(f"Boxplot quantitative variable {element}")
  plt.tight_layout()
```
